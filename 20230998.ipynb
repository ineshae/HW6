### 1) 리프 노드 개수
* **풀이**: 리프 노드는 자식이 없는 노드입니다.
* **대상**: 7, 17, 70
* **답**: **3개**

---

### 2) 높이 4에서 가능한 최대 리프 노드 수
* **풀이**: 포화 이진 트리에서 리프 노드 수는 $2^{h-1}$ 입니다.
* **계산**: $2^{4-1} = 2^3 = 8$
* **답**: **8개**

---

### 3) 리프 노드와 내부 노드 차이
* **풀이**:
    * 전체 노드 수: $2^4 - 1 = 15$
    * 리프 노드: $2^{4-1} = 8$
    * 내부 노드: $15 - 8 = 7$
* **차이**: $8 - 7 = 1$
* **답**: **1**

---

### 4) 일반화
* **성질**: 높이 $h$인 포화 이진 트리에서 리프 노드와 내부 노드의 차이는 항상 $1$입니다.
* **답**: **항상 성립한다**

## 02. 완전 이진 트리 성질

### 1) 리프 노드 수
* **공식**: $\lceil n/2 \rceil$

### 2) 내부 노드와의 관계
* **분석**:
    * 내부 노드 수: $\lfloor n/2 \rfloor$
    * 리프 노드 수: $\lceil n/2 \rceil$
* **결론**: 리프 노드는 내부 노드와 **같거나 1개 많습니다.**

## 03. 이진 트리 순회

### 1) 전위 순회 (Pre-order)
* **순서**: 루트 → 왼쪽 → 오른쪽
* **결과**: `5, 3, 3, 8, 7, 6, 9`

### 2) 중위 순회 (In-order)
* **순서**: 왼쪽 → 루트 → 오른쪽
* **결과**: `3, 3, 5, 6, 7, 8, 9`

### 3) 후위 순회 (Post-order)
* **순서**: 왼쪽 → 오른쪽 → 루트
* **결과**: `3, 3, 6, 7, 9, 8, 5`

### 4) 중위 순회 정렬 증명
* **설명**: 이진 검색 트리의 왼쪽은 루트보다 작고, 오른쪽은 루트보다 큽니다. 중위 순회는 이들을 크기 순으로 방문하므로 결과는 항상 **오름차순 정렬**됩니다.
